```python
# Step 1: Open first file
file1 = open('hello.txt', 'r')

# Step 2: Read content of each line and put it
# together in one line
full_file_content = ''
line_number = 0
for line in file1: 
    print ('Line read: ' + line)
    line_number = line_number + 1
    full_file_content = full_file_content + str(line_number) + ':' +  line.strip()
    print ('Full content till now: ' + full_file_content)
    
# Step 3: Close first file
file1.close()

# Step 4: Open second file to write to
file2 = open('hello2.txt', 'w')

# Step 5: Write full content to output file
file2.write(full_file_content)

# Step 6: Close second file
file2.close()
    
```
