```python
from pylab import plot, show, bar, scatter, annotate
x = []
y = []
file1 = open('txt.txt', 'r')
for line in file1:
    line = line.strip()
    line = line.split(',')
    x.append(int(line[0]))
    y.append(int(line[1]))
file1.close()
print (x)
print (y)
scatter(x,y)
show()
```
